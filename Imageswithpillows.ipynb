```python
# check PIL and Pillow version numbers
import PIL
from PIL import Image
print('Pillow Version:', PIL.__version__)
print('PIL Version:', PIL.__version__) # Corrected to use __version__
```

```python
#load the image
image = Image.open('Sydney-Opera-House.jpg')
#summarize some details about the image
print(image.format)
print(image.mode)
print(image.size)
image.show()
```

```python
#load and display an image using matplotlib
from matplotlib import image
from matplotlib import pyplot
```

```python
#load image as pixel array
data = image.imread('Sydney-Opera-House.jpg')
#summarize shape of the pixel array
print(data.dtype)
print(data.shape)
#display the array of pixels as an image
pyplot.imshow(data)
pyplot.show()
```

```python
#load image and covert to and from numpy array
from PIL import Image
from numpy import asarray
#load the image
image = Image.open('Sydney-Opera-House.jpg')
#convert image to numpy array
data = asarray(image)
#summarize array shape
print(data.shape)
#create Pillow image
image2 = Image.fromarray(data)
#summarize image details
print(image2.format)
print(image2.mode)
print(image2.size)
```

```python
#Example of saving an image in another format
from PIL import Image
image = Image.open('Sydney-Opera-House.jpg')
image.save('sydney.png', format='PNG')
image2 = Image.open('sydney.png')
print(image2.format)
```

```python
#converting an image to grayscale
from PIL import Image
image = Image.open('Sydney-Opera-House.jpg')
#covert the image
gsImage = image.convert(mode='L')
#save in jpg format
gsImage.save('sydney_gray.jpg')
#load the image again and show it
image2 = Image.open('sydney_gray.jpg')
image2.show()
```

```python
#How to resize images
from PIL import Image
image = Image.open('Sydney-Opera-House.jpg')
#report the size of the image
print(image.size)
#create a thumbnail and preserve aspect ratio
image.thumbnail((100,100))
#report the size of the modified image
print(image.size)
image.show()
```

```python
#resizing an image when we dont want to maintain the aspect ratio
from PIL import Image
image = Image.open('Sydney-Opera-House.jpg')
#report the size of the image
print(image.size)
#resize image
image_resized = image.resize((200,200))
print(image_resized.size)
```

```python
#Flip an image
from PIL import Image
from matplotlib import pyplot
image = Image.open('Sydney-Opera-House.jpg')
#horizontal flip
hoz_flip = image.transpose(Image.FLIP_LEFT_RIGHT)
#vertical flip
ver_flip = image.transpose(Image.FLIP_TOP_BOTTOM)
# plot all three images using matplotlib
pyplot.subplot(311)
pyplot.imshow(image)
pyplot.subplot(312)
pyplot.imshow(hoz_flip)
pyplot.subplot(313)
pyplot.imshow(ver_flip)
pyplot.show()
```

```python
#Rotate an image
from PIL import Image
from matplotlib import pyplot
image = Image.open('Sydney-Opera-House.jpg')
pyplot.subplot(311)
pyplot.imshow(image)
#rotate the image 45 degrees
pyplot.subplot(312)
pyplot.imshow(image.rotate(45))
#rotate 90 degrees
pyplot.subplot(313)
pyplot.imshow(image.rotate(90))
pyplot.show()
```
